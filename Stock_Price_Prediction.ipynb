## Yahoo Finance Stock

```python
import pandas as pd
import yfinance as yf
```

```python
# initializing the ticker class
sp500 = yf.Ticker("^GSPC")
```

```python
# history prices
sp500 = sp500.history(period="max")
```

```python
# Stock price predictor data
sp500
```

## Cleaning Stocket Market Data

```python
# Closing price
sp500.plot.line(y ="Close", use_index=True)
```

```python
# Deleting two columns
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
# Shows tomorrows price
sp500["Tomorrow"] = sp500['Close'].shift(-1)
```

```python
sp500
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
```

```python
# Target column shows the price goes high or low
sp500
```

```python
# Remove the data before 1990
sp500 = sp500.loc["1990-01-01": ].copy()
```

```python
sp500
```

## Train the model

```python
# Model training Using Random forest classifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# min_sample helps the decision tree data not to overfit
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=42)
```

```python
# All rows in taring set and last 100 rows in testing
train = sp500.iloc[:-100]
test = sp500.iloc[-100:]
```

```python
pred = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[pred], train["Target"])
```

```python
from sklearn.metrics import precision_score
# precision_score Mathes our Prediction to real data
```

```python
preds = model.predict(test[pred])
```

```python
# This output comes in series
preds = pd.Series(preds, index=test.index)
```

```python
preds
```

```python
precision_score(test["Target"],  preds)
```

```python
# combine the actual to predicted value
combined = pd.concat([test["Target"], preds], axis=1)
```

```python
combined.plot()
```

```python
# Combined the actual and predicting Values
# hyperparamater tuning
def predict(train, test, pred, model):
    model.fit(train[pred], train["Target"])
    preds = model.predict(test[pred])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
# Taking 10 years of data 2500 1 year (250)
def backtest(data, model, pred, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train=data.iloc[0:i].copy()
        test=data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, pred, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, pred)
```

```python
predictions["Predictions"].value_counts()
```

```python
# 53% Accuracy of our predictions were correct and market goes up
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
# Percentage of days when market goes up
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
horizons = [2,5,60,250,1000]
```

```python
new_predictors = []
# This loop shows the predictions of last 2 days 5 days and so on
for horizon in horizons:
    rolling_avg = sp500.rolling(horizon).mean()

    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_avg["Close"]
# This will show the sum of the last 5 days of the target of stock price went up
    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]

    new_predictors+= [ratio_column, trend_column]
```

```python
# The NAN shows where it cant predict the rolling sum as it cant find the data before that day
sp500
```

```python
# removing the data with extra rows
sp500 = sp500.dropna()
```

```python
# To compute the data we should have the last years data
sp500
```

```python
# improving the model
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=42)
# comnone and finds the probability of predicton goes upor down
# model.predict_proba
def predict(train, test, pred, model):
    model.fit(train[pred], train["Target"])
    preds = model.predict_proba(test[pred])[:,1]
    # if it goes up then 60 it will be 1 and below it will be 0
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
# Now the model shows the prections more accurate when the predictions goes up or down
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions['Predictions'])
```
